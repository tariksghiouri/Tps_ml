```python
bitcoin=pd.read_csv('BTC-USD.csv', index_col='Date', parse_dates=True)
bitcoin
```

```python
bitcoin.index
```

```python
bitcoin['2022']['Close'].plot()
```

```python
bitcoin['2022-09']['Close'].plot()
```

```python
bitcoin.loc['2022','Close'].resample('W').plot()
```

```python
bitcoin.loc['2022','Close'].resample('w').mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2022','Close'].plot()
bitcoin.loc['2022','Close'].resample('M').mean().plot()
bitcoin.loc['2022','Close'].resample('M').std().plot()
plt.legend()
plt.show()
```

```python
bitcoin['Close'].resample('W').agg(['mean','std'])
```

```python
m=bitcoin['Close'].resample('W').agg(['mean','std','min','max'])
plt.figure(figsize=(12,8))
m['mean']['2022'].plot(label='m par sem')
plt.fill_between(m.index,m['max'],m['min'],alpha=.2, label='min max paer sem')
plt.legend()
plt.show()
```

```python
bitcoin.loc['2022','Close'].rolling(window=7)
```

```python
bitcoin.loc['2022','Close'].rolling(window=7).mean().plot()
```

```python
bitcoin.loc['2022-07','Close'].rolling(window=7).mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2022-06','Close'].plot()
bitcoin.loc['2022-06','Close'].rolling(window=7).mean().plot(label='mon centre',lw=3,ls=':',alpha=0)
bitcoin.loc['2022-06','Close'].plot()
```
